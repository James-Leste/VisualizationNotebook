```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Let's generate more structured fake data for 7 countries, volumes in the range of 0 to 5000.
np.random.seed(0)  # For reproducible results
years = np.arange(1994, 2006)

# Create a base volume for each country that increases each year, to ensure separation between countries
base_volumes = np.linspace(500, 4500, num=7)

# Data will vary around the base volumes with some noise
data = np.array([np.random.normal(base, 250, len(years)) for base in base_volumes]).T

# Ensure that the data is within the specified limits (0-5000)
data[data < 0] = 0
data[data > 5000] = 5000

# Creating a DataFrame with more structured data
df_structured = pd.DataFrame(data, index=years, columns=['Ecuador', 'Philippines', 'Costa Rica', 'Colombia', 'Guatemala', 'Honduras', 'Panama'])

# Plotting the lineplot with the structured data
plt.figure(figsize=(14, 8))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

for i, country in enumerate(df_structured.columns):
    plt.plot(df_structured.index, df_structured[country], marker='o', label=country, color=colors[i % len(colors)])

plt.title('Export of Bananas in Metric Tons (1994-2005)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Export Volume (in tons)', fontsize=12)
plt.legend(loc="upper left", ncol=7, fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.savefig("../images/banana.png")
plt.show()
```
